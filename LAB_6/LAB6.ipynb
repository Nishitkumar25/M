```python
def summation_unit(inputs, weights):
    return sum(i * w for i, w in zip(inputs, weights))
```

```python
def step_act(x):
    return 1 if x >= 0 else 0

def bipolar_step_act(x):
    return 1 if x >= 0 else -1

def sigmoid_act(x):
    return 1 / (1 + np.exp(-x))

def tanh_act(x):
    return math.tanh(x)

def relu_act(x):
    return max(0, x)

def leaky_relu_act(x, alpha=0.01):
    return x if x > 0 else alpha * x
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
def act(val,n):
    if n == 0:
        return step_act(val)
    elif n == 1:
        return bipolar_step_act(val)
    elif n == 2:
        return sigmoid_act(val)
    elif n == 3:
        return tanh_act(val)
    elif n == 4:
        return relu_act(val)
    elif n == 5:
        return leaky_relu_act(val)
    else:
        return None
```

```python
# Derivative of the Sigmoid function (for backpropagation)
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Inputs
y = np.array([[0], [0], [0], [1]])  # Outputs (AND gate)

# Initialize weights and biases randomly
np.random.seed(42)
weights_input_hidden = np.random.rand(2, 2)  
weights_hidden_output = np.random.rand(2, 1) 
bias_hidden = np.random.rand(1, 2)  
bias_output = np.random.rand(1, 1)  

# Learning rate
alpha = 0.05
final_output = None

epochs = 10000
for epoch in range(epochs):
    #forward pass
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden  # Input to hidden layer
    hidden_output = sigmoid_act(hidden_input)  # Output from hidden layer
    
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output  # Input to output layer
    final_output = sigmoid_act(final_input)  # Final output (prediction)
    
    #mean squared error
    error = y - final_output
    
    # Backpropagation
    # Output layer error gradient
    d_output = error * sigmoid_derivative(final_output)
    
    # Hidden layer error gradient
    hidden_error = d_output.dot(weights_hidden_output.T)
    d_hidden = hidden_error * sigmoid_derivative(hidden_output)
    
    # Update weights and biases
    weights_hidden_output += hidden_output.T.dot(d_output) * alpha
    weights_input_hidden += X.T.dot(d_hidden) * alpha
    bias_output += np.sum(d_output, axis=0, keepdims=True) * alpha
    bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * alpha

# After training, let's see the final output
print("Final output after training:")
print(final_output)

# Test predictions after training
def predict(X):
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid_act(hidden_input)
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    return sigmoid_act(final_input)

# Predictions for the AND gate
predictions = predict(X)
print("\nPredictions for AND gate logic (rounded):")
print(np.round(predictions))  # Rounding the output for binary classification
```
